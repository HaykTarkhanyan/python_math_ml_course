```python
import random
import string

random.seed(509)

letters_and_digits_and_space = string.ascii_letters + string.digits + " "
print(letters_and_digits_and_space)

password = "hndkahav"
pass_len = len(password)
```

```python
POPULATION_SIZE = 1_000

population = []

for _ in range(POPULATION_SIZE):
    word = random.sample(letters_and_digits_and_space,
                         k=pass_len)
    population.append("".join(word))
```

```python
def fitness(x):
    score = 0 
    
    for i in range(pass_len):
        if x[i] == password[i]:
            score += 1
            
    return score
```

```python
scores = [fitness(i) for i in population]
```

```python
import numpy as np

biggest_index = np.argmax(scores)

mother = population[biggest_index]

scores.pop(biggest_index)
population.pop(biggest_index)

second_biggest_index = np.argmax(scores)
father = population[second_biggest_index]

print(mother)
```

```python
parent = father

parent_score = fitness(parent)
random_indexes = random.sample(range(pass_len), k=parent_score)
print(random_indexes)

new_member = ""

for i in range(pass_len):
    if i in random_indexes:
        new_member += parent[i]
    else:
        new_member += random.choice(letters_and_digits_and_space)

print(parent) 
print(new_member)
```

```python
letters_and_digits_and_space = string.ascii_letters + string.digits + " "
MAX_ITER = 1_000
POPULATION_SIZE = 1_000
MUTATION_PROBABILITY = 0.1


def genetic_algorithm(password, max_iterations = 100, POPULATION_SIZE=1000, MUTATION_PROBABILITY=0.1):
    pass_len = len(password)

    population = []

    for _ in range(POPULATION_SIZE):
        word = random.sample(letters_and_digits_and_space,
                            k=pass_len)
        population.append("".join(word))
        
    for iteration in range(max_iterations):
        scores = [fitness(i) for i in population]

        biggest_index = np.argmax(scores)

        mother = population[biggest_index]
        print(f"{iteration}: Best score = {scores[biggest_index]}")

        if fitness(mother) == pass_len:
            print(f"Password found: {mother} in {iteration} iterations")
            return mother

        scores.pop(biggest_index)
        population.pop(biggest_index)

        second_biggest_index = np.argmax(scores)
        father = population[second_biggest_index]

        new_population = [mother, father]

        while len(new_population) < POPULATION_SIZE:
            parent = random.choice([mother, father])

            parent_score = fitness(parent)
            random_indexes = random.sample(range(pass_len), k=parent_score)

            new_member = ""

            for i in range(pass_len):
                if i in random_indexes:
                    new_member += parent[i]
                else:
                    new_member += random.choice(letters_and_digits_and_space)

            new_member_mutated = ""

            for i in range(pass_len):
                if random.random() < MUTATION_PROBABILITY:
                    new_member_mutated += random.choice(letters_and_digits_and_space)
                else:
                    new_member_mutated += new_member[i]
                    
            new_population.append(new_member_mutated)
        
        population = new_population
```

```python
genetic_algorithm(password, max_iterations=MAX_ITER, MUTATION_PROBABILITY=0.1, POPULATION_SIZE=10) 
```
