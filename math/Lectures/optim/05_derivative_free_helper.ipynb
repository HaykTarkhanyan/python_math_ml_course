```python
import numpy as np

import plotly.graph_objects as go

# Create grid
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z = |x+y| + 3|y-x|
Z = np.abs(X + Y) + 3 * np.abs(Y - X)

# Create 3D surface plot
fig_3d = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])
fig_3d.update_layout(
    title='3D Plot of |x+y| + 3|y-x|',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z'
    ),
    width=800,
    height=800
)
fig_3d.show()

# Create contour plot
fig_contour = go.Figure(data=[go.Contour(x=x, y=y, z=Z, colorscale='Viridis')])
fig_contour.update_layout(
    title='Contour Plot of |x+y| + 3|y-x|',
    xaxis_title='x',
    yaxis_title='y',
    width=800,
    height=800
)
fig_contour.show()
```

```python
# Extract values along the line y = -5
y_line = -5
y_idx = np.argmin(np.abs(y - y_line))
z_line = Z[y_idx, :]

# Create a line plot showing z values along y = -5
fig_line = go.Figure()
fig_line.add_trace(go.Scatter(x=x, y=z_line, mode='lines', name=f'z at y={y[y_idx]:.2f}'))
fig_line.update_layout(
    title=f'Values of |x+y| + 3|y-x| along line y = {y[y_idx]:.2f}',
    xaxis_title='x',
    yaxis_title='z',
    width=800,
    height=500
)
fig_line.show()
```

# Levy function

```python
# Calculate Levy function: sin(3x) + (x-1)[1+sin(3x)] + (x-1)[1+sin(2x)]
Z_levy = (np.sin(3 * np.pi * X)**2 + 
          (X - 1)**2 * (1 + np.sin(3 * np.pi * Y)**2) + 
          (Y - 1)**2 * (1 + np.sin(2 * np.pi * Y)**2))


# Create 3D surface plot for Levy function
fig_levy_3d = go.Figure(data=[go.Surface(x=X, y=Y, z=Z_levy, colorscale='Viridis')])
fig_levy_3d.update_layout(
    title='3D Plot of Levy Function',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z'
    ),
    width=800,
    height=800
)
fig_levy_3d.show()
```

```python

# Create contour plot for Levy function
fig_levy_contour = go.Figure(data=[go.Contour(x=x, y=y, z=Z_levy, colorscale='Viridis')])
fig_levy_contour.update_layout(
    title='Contour Plot of Levy Function',
    xaxis_title='x',
    yaxis_title='y',
    width=800,
    height=800
)
fig_levy_contour.show()
```
