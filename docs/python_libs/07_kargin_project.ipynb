```python
from fuzzywuzzy import fuzz
```

```python
def get_similarity(text_kargin, search_text):
    return fuzz.token_sort_ratio(search_text.lower(), text_kargin)
```

```python
df['similaratiy'] = df["text"].apply(lambda x: get_similarity(x, search_text))
```

```python
df.sort_values("similaratiy", ascending=False)
```

## Youtube-ีซึ ีซีถึีธ

```python
from pytubefix import YouTube 
```

```python
def get_video_info(video_id):
    yt = YouTube(video_id)
    return {
        "title": yt.title,
        "views": yt.views,
        "length": yt.length,
        "author": yt.author,
        "published_at": yt.publish_date
    }
```

```python
df_small = df.head()
```

```python
df_small["youtube_info"] = df_small["links"].apply(get_video_info)
```
