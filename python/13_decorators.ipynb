## first class object

ֆունցկաիները կոչվում են First-Class Objectներ, ինչը նշանակումա որ իրանց կարանք որպես արգումենտ տանք ուրիշ ֆունկցիաներին,
ուրիշ ֆունկցիաները կարան վերադարձնեն ֆունկցիաներ, intի նման ասենք


```python
def ispanakan(anun):
    return f"Բարև {anun}իտո"

def dinozavrakan(anun):
    return f"Բարև {anun}ազավռ"

def barevel(anun, specific_barev):
    return specific_barev(anun)

anun = 'Հայկ'
print(barevel(anun, dinozavrakan))
```

## վերադարձ ունեցող ֆունկցիայի օրինակ

```python
def divide(x, y):
    return x / y

def multiply(x, y):
    return x * y
```

```python
print(divide(22, 7))
print(multiply(10.5, 11.509))
```

```python
# ուզում ենք կլորացնենք
```

```python
def divide(x, y):
    return_value = x / y
    
    return round(return_value, 2)

def multiply(x, y):
    return_value = x * y
    
    return round(return_value, 2)
```

```python
print(divide(22, 7))
print(multiply(10.5, 11.509))
```

```python
# եկեք բայց դեկորատորներով անենք
```

```python
def rounder(func):
    def wrapper(x, y): # գրկող
        res = func(x, y)
        return round(res, 2)
    return wrapper
```

```python
@rounder
def divide(x, y):
    return x / y

@rounder
def multiply(x, y):
    return x * y
```

```python
print(divide(22, 7))
print(multiply(10.5, 11.509))
```

```python
def timer(func):
    def nor_func(*args, **kwargs):
        print("starting")
        st = time.perf_counter()
        
        res = func(*args, **kwargs)
        duration = time.perf_counter() - st
        
        print("done")
        print(duration)
        return res

    return nor_func
```

```python
def rounder(func):
    def wrapper(*args, **kwargs): # գրկող
        res = func(*args, **kwargs)
        return round(res, 2)
    return wrapper
```

```python
@timer
@rounder
def divide(x, y):
    return x / y

@timer
@rounder
def multiply(x, y):
    return x * y

# timer(rounder(multiply))
```

```python
def erku_angam(func):
    def nor_func(*args, **kwargs):
        print("arajin angam")
        func(*args, **kwargs)
        
        print("erku angam")
        func(*args, **kwargs)
    return nor_func
```

```python
@erku_angam
def divide(x, y):
    print(x / y)

divide(3,4)
```

## մի քանի օրինակ

### ֆունկցիայի դանդաղացուցիչ

```python
def mi_vrazi(func):
    def wrapper_func(*args, **kwargs):
        time.sleep(1)
        return func(*args, **kwargs)
    return wrapper_func
```

```python
def scrape(url):
    print(f"քաշեցի {url}ի տվյալները")
```

```python
urls = ['https://www.instructables.com/Basic-Steps-of-How-to-Make-Cheese/', 'https://www.youtube.com/watch?v=8DYyjqpZRlQ'] * 3

for url in urls:
    scrape(url)
```

```python
@mi_vrazi
def scrape(url):
    print(f"քաշեցի {url}ի տվյալները")
```

## debuger

```python
from datetime import datetime

def debug(func):
    def wrapper(*args, **kwargs):
        print(f"մուտքում ստացել եմ {args} և {kwargs} արգումենտները")
        print(f"Սկսելու ժամանակը {datetime.now().strftime('%H:%M:%S')}")
        st = time.perf_counter()
        
        res = func(*args, **kwargs)
        
        end = time.perf_counter()
        print(f"ծրագիրը աշխատեց {round(end-st, 2)} վարկյանում")
        print(f"ծրագիրը վերադարձրեց՝ {res}\n")
        return res
    return wrapper


```

```python
@debug
@mi_vrazi
def scrape(url):
    print(f"քաշեցի {url}ի տվյալները")
```

```python
urls = ['https://www.instructables.com/Basic-Steps-of-How-to-Make-Cheese/', 'https://www.youtube.com/watch?v=8DYyjqpZRlQ'] * 3

for url in urls:
    scrape(url)
```

# Functools example

```python
from functools import lru_cache # Least Recently Used

def gcd(a, b): # greatest common divisor
    time.sleep(1)
    
    while b != 0: # էվկիդլեսի ալգորիթմ
        a, b = b, a % b
    return a

```

```python
print(gcd(10, 15))
```

```python
%%time
print(gcd(11_509_000_000, 11_509_000_0001))
```

```python
from functools import lru_cache

@lru_cache
def gcd(a, b):
    time.sleep(1)
    while b != 0:
        a, b = b, a % b
    return a

```

```python
%%time
print(gcd(11_509_000_000, 11_509_000_0001))
```

```python
%%time
print(gcd(11_509_000_000, 11_509_000_0001))
```

```python
%%time
print(gcd(11_509_000_001, 11_509_000_002))
```
