### ‘≤’°’¶’¥’∏÷Ç’©’µ’∏÷Ç’∂’∂’•÷Ä’´ ’∞’•’ø ’£’∏÷Ä’Æ’∏’≤’∏÷Ç’©’µ’∏÷Ç’∂’∂’•÷Ä

image.png

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# ’Ñ’´’°’æ’∏÷Ä’∏÷Ç’¥
print(set1.union(set2)) # Output: {1, 2, 3, 4, 5, 6}
# ’Ø’°’¥ ’ß’¨
# set1 | set2 (’Ø’°’¥) # caps lock + right shift-’´ ’Ø’∏’≤÷Ñ’® 6

# ’Ä’°’ø’∏÷Ç’¥
print(set1.intersection(set2)) # Output: {3, 4}
# ’Ø’°’¥ ’ß’¨
# set1 & set2 (÷á) # uppercase 6

# ’è’°÷Ä’¢’•÷Ä’∏÷Ç’©’µ’∏÷Ç’∂
print(set1.difference(set2)) # Output: {1, 2}
# ’Ø’°’¥ ’ß’¨
# print(set1 - set2)

# ’ç’´’¥’•’ø÷Ä’´’Ø ’ø’°÷Ä’¢’•÷Ä’∏÷Ç’©’µ’∏÷Ç’∂
print(set1.symmetric_difference(set2)) # Output: {1, 2, 5, 6}
# ’Ø’°’¥ ’ß’¨
# set1 ^ set2 # uppercase 6
```

```python
import json 

PATH = "result.json"

with open(PATH, "r") as f:
    data = json.load(f)
```

```python
chats = data["chats"]["list"]

chats[0]["type"]
```

```python
types = [i["type"] for i in chats]

count_dict = {i:0 for i in types}

for i in types:
    count_dict[i] += 1
    
count_dict
```

```python
count_has_message = 0
chats_not_empty = []

for chat in chats:
    if chat["messages"]:
        count_has_message += 1 
        chats_not_empty.append(chat["name"])

print(f"Perc - {count_has_message / len(chats) * 100:.2f}")
print(chats_not_empty)
```

```python
chats_list_not_empty = [chat for chat in chats if chat["messages"]]
```

```python
msg_counts = {}

for chat in chats_list_not_empty:
    count = 0
    
    msgs = chat["messages"]
    for msg in msgs:
        if msg["type"] == "message":
            count += 1
    
    msg_counts[chat["name"]] = count
                
print(msg_counts)
        # if chat["name"][""]
        # print(*chat["messages"][10:15], sep="\n")
```
